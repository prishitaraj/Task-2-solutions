```python
import pandas as pd
```

```python
df=pd.read_csv('2019_nCoV_20200121_20200127.csv')
```

```python
df
```

```python
df.shape
```

```python
df.dropna(axis='columns', how='any')
```

```python
m1= df['Confirmed'].mean()
m2= df['Suspected'].mean()
m3= df['Recovered'].mean()
m4= df['Death'].mean()
```

```python
df['Confirmed'].fillna(m1, inplace=True)
df['Suspected'].fillna(m2, inplace=True)
df['Recovered'].fillna(m3, inplace=True)
df['Death'].fillna(m4, inplace=True)
```

```python
df.to_csv ('new dataframe.csv')
```
