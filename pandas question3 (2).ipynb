```python
import pandas as pd
```

```python
df=pd.read_csv('2019_nCoV_20200121_20200127.csv')
```

```python
df
```

```python
filt=input("Enter the column name to be sorted- ")
if(filt =='Province/State'):
    df.sort_values(by='Province/State', inplace=True)
if(filt=='Country/Region'):
    df.sort_values(by='Country/Region', inplace=True)
if(filt=='Confirmed'):
    df.sort_values(by='Confirmed', inplace=True)
```
